```python
def download_images(search_url, folder_path="./remote_images"):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    response = requests.get(search_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    img_tags = soup.find_all('img')
    count = 0

    for img in img_tags:
        if count >= 100:
            break
        
        if 'src' not in img.attrs:
            continue
        img_url = img['src']
        
        if(img_url.find(".jpg") == -1):
            continue
        
        full_img_url = urljoin(search_url, img_url)
        
        img_name = os.path.basename(full_img_url)
        
        img_path = os.path.join(folder_path, img_name)
        
        try:
            with open(img_path, 'wb') as f:
                img_data = requests.get(full_img_url).content
                f.write(img_data)
            print(f"Downloaded {img_name}")
            count += 1
        except Exception as e:
            print(f"Failed to download {img_name}. Reason: {e}")
```

```python
def make_collage(image_folder, collage_name='./collage.jpg', images_per_row=15):
    image_paths = [os.path.join(image_folder, img) for img in os.listdir(image_folder) if img.endswith(('.png', '.jpg', '.jpeg'))]
    
    images = [Image.open(img_path) for img_path in image_paths]
    widths, heights = zip(*(i.size for i in images))

    max_width = max(widths)
    max_height = max(heights)

    # Size of the grid
    num_images = len(images)
    grid_width = images_per_row * (max_width) 
    grid_height = ((num_images // images_per_row) + (num_images % images_per_row > 0)) * (max_height)

    # Create a new image with a white background
    collage_image = Image.new('RGB', (grid_width, grid_height), color='white')

    # Paste images into collage
    for index, image in enumerate(images):
        row = index // images_per_row
        col = index % images_per_row
        collage_image.paste(image, (col * max_width, row * max_height))

    collage_image.save(collage_name)
    print(f"Collage saved as {collage_name}")
    print(grid_width, grid_height)
```

```python
def rename_files_in_directory(directory_path, prefix):
    files = os.listdir(directory_path)
    for index, file_name in enumerate(files, start=1):
        new_file_name = f"{prefix}_{index}{os.path.splitext(file_name)[1]}"
        old_file_path = os.path.join(directory_path, file_name)
        new_file_path = os.path.join(directory_path, new_file_name)
        os.rename(old_file_path, new_file_path)
        print(f"Renamed {file_name} to {new_file_name}")
rename_files_in_directory("./mobile_images", "mobile")
rename_files_in_directory("./remote_images", "remote")
```

```python
def resize_and_pad_images(directory_path, output_size):
    for file_name in os.listdir(directory_path):
        file_path = os.path.join(directory_path, file_name)
        image = Image.open(file_path)
        ratio = min(output_size[0] / image.size[0], output_size[1] / image.size[1])
        new_size = (int(image.size[0] * ratio), int(image.size[1] * ratio))
        image = image.resize(new_size, Image.ANTIALIAS)

        new_image = Image.new("RGB", output_size, "white")
        new_image.paste(image, ((output_size[0] - new_size[0]) // 2,
                                (output_size[1] - new_size[1]) // 2))
        new_image.save(file_path)
        print(f"Processed {file_name}")
```

DATA AUGMENTATION

```python
def augment_images(image_directory, output_directory, prefix, final_target):
    # Ensure the output directory exists
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Initialize the ImageDataGenerator with desired settings
    datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.4,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    # Repeat the augmentation loop until the total number of images is reached
    current_total = len(os.listdir(image_directory))
    while current_total < final_target:
        original_images = [img for img in os.listdir(image_directory) if img.lower().endswith(('.png', '.jpg', '.jpeg'))]
        for file_name in original_images:
            if current_total >= final_target:
                break  # Exit if the target is reached during the loop

            img_path = os.path.join(image_directory, file_name)
            img = load_img(img_path)  # Load the image file
            img_array = img_to_array(img)  # Convert it to a numpy array
            img_array = img_array.reshape((1,) + img_array.shape)  # Reshape it

            # Generate one new image
            for batch in datagen.flow(img_array, batch_size=1):
                new_image = Image.fromarray(batch[0].astype('uint8'), 'RGB')
                new_file_name = f"{prefix}_{current_total + 1}.jpeg"
                new_image.save(os.path.join(output_directory, new_file_name))
                current_total += 1
                break  # Only generate one image per original image per cycle

        print(f"Updated total: {current_total} images")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.applications import VGG16
from keras import layers, models, optimizers
```

```python
# Initialize the VGG16 base with pre-trained weights.
conv_base = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
# Freeze the convolutional base to prevent weights from being updated during the first phase of training.
conv_base.trainable = False
```

```python
# Preprocessing the data - only rescaling the images.
datagen = ImageDataGenerator(rescale=1./255)
```

Since we have separate folders for each class and train/validation, we need to combine them in a single directory structure that can be used by `flow_from_directory`. Keras requires a directory per class, and under each class directory, there should be 'train' and 'validation' directories.

```python
# Create the data generators for training and validation sets.
train_generator = datagen.flow_from_directory(
    "./train_dir",  # This is the path to the directory where train_mobile_dir and train_remote_dir are subdirectories.
    target_size=(224, 224),
    batch_size=20,
    class_mode='binary')

validation_generator = datagen.flow_from_directory(
    "./validation_dir",  # This is the path to the directory where validation_mobile_dir and validation_remote_dir are subdirectories.
    target_size=(224, 224),
    batch_size=20,
    class_mode='binary')
```

```python
# Create the model and add new dense layers for binary classification.
full_model = models.Sequential()
full_model.add(conv_base)
full_model.add(layers.Flatten())
full_model.add(layers.Dense(256, activation='relu'))
full_model.add(layers.Dropout(0.5))  # Adding dropout for regularization
full_model.add(layers.Dense(1, activation='sigmoid'))  # Since we're doing binary classification
```

```python
# Compile the model.
full_model.compile(optimizer=optimizers.RMSprop(learning_rate=2e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model.
fullModel_history = full_model.fit(
    train_generator,
    steps_per_epoch=50,  # Adjust based on the size of your dataset.
    epochs=30,
    validation_data=validation_generator,
    validation_steps=50  # Adjust based on the size of your dataset.
)
```

```python
# print(history.params)

# check the keys of history object
print(fullModel_history.history['accuracy'])
```

```python
# Plot training & validation accuracy values
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(fullModel_history.history['accuracy'])
plt.plot(fullModel_history.history['val_accuracy'])
plt.title('VGG16 Model accuracy with 1200 image samples from each class')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
```

```python
# Plot training & validation loss values
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.plot(fullModel_history.history['loss'])
plt.plot(fullModel_history.history['val_loss'])
plt.title('VGG16 Model loss with 1200 images from each class')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()

plt.show()
```

```python
# Load the ResNet50 base model, pre-trained on ImageNet
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the convolutional base
base_model.trainable = False

# Preprocessing the data
train_datagen = ImageDataGenerator(rescale=1./255)
validation_datagen = ImageDataGenerator(rescale=1./255)

# Create the data generators for training and validation sets
train_generator = train_datagen.flow_from_directory(
    './set1-300/train_dir',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary')  # or 'categorical' for multiclass

validation_generator = validation_datagen.flow_from_directory(
    './set1-300/validation_dir',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary')  # or 'categorical' for multiclass

# Create the model
resnetModel1 = models.Sequential()
resnetModel1.add(base_model)
resnetModel1.add(layers.GlobalAveragePooling2D())
resnetModel1.add(layers.Dense(256, activation='relu'))
resnetModel1.add(layers.Dropout(0.5))
resnetModel1.add(layers.Dense(1, activation='sigmoid'))  # Use 'softmax' for multiclass

# Compile the model
resnetModel1.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
              loss='binary_crossentropy',  # Use 'categorical_crossentropy' for multiclass
              metrics=['accuracy'])

# Train the model
resnethistory1 = resnetModel1.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=20,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size)
```

```python
# Plot training & validation accuracy values

import pandas as pd
import matplotlib.pyplot as plt

# Convert the history.history dict to a pandas DataFrame
resnethistory1_df = pd.DataFrame(resnethistory1.history)

# Filter out the epochs where the accuracy is zero
filtered_history_df = resnethistory1_df[(resnethistory1_df['accuracy'] != 0) & (resnethistory1_df['accuracy'] != 0)]

# Plot training & validation loss values
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.plot(filtered_history_df['accuracy'])
plt.plot(filtered_history_df['val_accuracy'])
plt.title('Resnet50 Model Accuracy with 300 samples from each class')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()

plt.show()
```

```python
# Plot training & validation accuracy values

# Filter out the epochs where the accuracy is zero
filtered_history_df = resnethistory1_df[(resnethistory1_df['loss'] != 0) & (resnethistory1_df['val_loss'] != 0)]

# Plot training & validation loss values
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.plot(filtered_history_df['loss'])
plt.plot(filtered_history_df['val_loss'])
plt.title('Resnet50 Model loss with 300 samples from each class')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()

plt.show()
```

# Resnet50 with all dataset sizes and 300 validation samples

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import models, layers, optimizers
```

```python
from keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import GlobalAveragePooling2D, Dense, Dropout
from keras import optimizers
import numpy as np

# Function to create and train a model
def train_model(pretrained=True, train_dir='./set1-300/train_dir', val_dir='./validation_dir'):
    if pretrained:
        base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
        base_model.trainable = False
    else:
        base_model = ResNet50(weights=None, include_top=False, input_shape=(224, 224, 3))
        base_model.trainable = True  # Train the whole model if no pretrained weights are used

    # Data generators
    train_datagen = ImageDataGenerator(rescale=1./255)
    validation_datagen = ImageDataGenerator(rescale=1./255)

    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

    validation_generator = validation_datagen.flow_from_directory(
        val_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

    # Building the model
    model = Sequential()
    model.add(base_model)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=20,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size)

    # Capture the best validation accuracy
    best_val_acc = max(history.history['val_accuracy'])
    return best_val_acc

# Training sizes and directories setup
training_sizes = [300, 500, 700, 1000, 1200]
training_dirs = ['./set1-300/train_dir', './set2-500/train_dir', './set3-700/train_dir', './set4-1000/train_dir', './train_dir']
validation_dir = './validation_dir'  # Fixed validation directory
resnet50_pretrained = []
resnet50_baseline = []

# Train models for each size
for size, dir in zip(training_sizes, training_dirs):
    print(f"Training pretrained model with {size} samples...")
    val_acc_pretrained = train_model(pretrained=True, train_dir=dir, val_dir=validation_dir) 
    resnet50_pretrained.append(val_acc_pretrained)

    print(f"Training baseline model with {size} samples...")
    val_acc_baseline = train_model(pretrained=False, train_dir=dir, val_dir=validation_dir)
    resnet50_baseline.append(val_acc_baseline)
    
    print("\npretrained array: ", resnet50_pretrained, "\nbaseline array: ", resnet50_baseline)
```

# VGG16 with all dataset sizes and 300 validation samples

```python
from keras.applications import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout
from keras import optimizers
import numpy as np

# Function to create and train a model
def train_vgg16_model(pretrained=True, train_dir = './set1-300/train_dir', val_dir='./validation_dir', training_size = 300):
    if pretrained:
        conv_base = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
        conv_base.trainable = False
    else:
        conv_base = VGG16(weights=None, include_top=False, input_shape=(224, 224, 3))
        conv_base.trainable = True  # Train the whole model if no pretrained weights are used

    # Preprocessing the data - only rescaling the images.
    datagen = ImageDataGenerator(rescale=1./255)
    
    # Data generators
    train_generator = datagen.flow_from_directory(
        train_dir,
        target_size=(224, 224),
        batch_size=20,
        class_mode='binary')

    validation_generator = datagen.flow_from_directory(
        val_dir,
        target_size=(224, 224),
        batch_size=20,
        class_mode='binary')

    # Create the model and add new dense layers for binary classification.
    model = Sequential()
    model.add(conv_base)
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(optimizer=optimizers.RMSprop(learning_rate=2e-5),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=20,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size)
    
    # Return the best validation accuracy
    return max(history.history['val_accuracy'])

# Training sizes and directories setup
training_sizes = [300, 500, 700, 1000, 1200]
training_dirs = ['./set1-300/train_dir', './set2-500/train_dir', './set3-700/train_dir', './set4-1000/train_dir', './train_dir']
validation_dir = './validation_dir'  # Fixed validation directory
vgg16_pretrained = []
vgg16_baseline = []

# Train models for each size
for size, dir in zip(training_sizes, training_dirs):
    print(f"Training pretrained VGG16 model with {size} samples...")
    val_acc_pretrained = train_vgg16_model(pretrained=True, train_dir=dir, val_dir=validation_dir, training_size = size)
    vgg16_pretrained.append(val_acc_pretrained)

    print(f"Training baseline VGG16 model with {size} samples...")
    val_acc_baseline = train_vgg16_model(pretrained=False, train_dir=dir, val_dir=validation_dir, training_size = size)
    vgg16_baseline.append(val_acc_baseline)

    print("\npretrained array: ", vgg16_pretrained, "\nbaseline array: ", vgg16_baseline)
```

# MobileNet Model

```python
from keras.applications import MobileNet
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import GlobalAveragePooling2D, Dense, Dropout
from keras import optimizers


def train_mobilenet_model(pretrained=True, train_dir='./train_dir', val_dir='./validation_dir', training_size=300):
    # Load the MobileNet model
    if pretrained:
        base_model = MobileNet(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
        base_model.trainable = False  # Freeze the layers
    else:
        base_model = MobileNet(weights=None, include_top=False, input_shape=(224, 224, 3))
        base_model.trainable = True  # Train the whole model

    # Data preprocessing
    datagen = ImageDataGenerator(rescale=1./255)
    train_generator = datagen.flow_from_directory(
        train_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')
    validation_generator = datagen.flow_from_directory(
        val_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

    # Create the model
    model = Sequential()
    model.add(base_model)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification

    # Compile the model
    model.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=20,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size)

    return max(history.history['val_accuracy'])


training_sizes = [300, 500, 700, 1000, 1200]
training_dirs = ['./set1-300/train_dir', './set2-500/train_dir', './set3-700/train_dir', './set4-1000/train_dir', './train_dir']
validation_dir = './validation_dir'
mobilenet_pretrained = []
mobilenet_baseline = []

for size, dir in zip(training_sizes, training_dirs):
    print(f"Training pretrained MobileNet model with {size} samples...")
    mobilenet_pretrained.append(train_mobilenet_model(True, dir, validation_dir, size))
    
    print(f"Training baseline MobileNet model with {size} samples...")
    mobilenet_baseline.append(train_mobilenet_model(False, dir, validation_dir, size))
```

# Resnet50 on smaller datasets

```python
from keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import GlobalAveragePooling2D, Dense, Dropout
from keras import optimizers
import numpy as np

# Function to create and train a model
def resnet_train_model(pretrained=True, train_dir='./setA/train_dir', val_dir='./validation_dir'):
    if pretrained:
        base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
        base_model.trainable = False
    else:
        base_model = ResNet50(weights=None, include_top=False, input_shape=(224, 224, 3))
        base_model.trainable = True  # Train the whole model if no pretrained weights are used

    # Data generators
    train_datagen = ImageDataGenerator(rescale=1./255)
    validation_datagen = ImageDataGenerator(rescale=1./255)

    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

    validation_generator = validation_datagen.flow_from_directory(
        val_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

    # Building the model
    model = Sequential()
    model.add(base_model)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(optimizer=optimizers.Adam(learning_rate=0.0001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=20,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size)

    # Capture the best validation accuracy
    best_val_acc = max(history.history['val_accuracy'])
    return best_val_acc

# Training sizes and directories setup
new_training_sizes = [20, 50, 90, 120, 180, 250]
new_training_dirs = ['./setA/train_dir', './setB/train_dir', './setC/train_dir', './setD/train_dir', './setE/train_dir', './setF/train_dir']
validation_dir = './validation_dir'  # Fixed validation directory
new_resnet50_pretrained = []
new_resnet50_baseline = []

# Train models for each size
for size, dir in zip(new_training_sizes, new_training_dirs):
    print(f"Training pretrained model with {size} samples...")
    val_acc_pretrained = resnet_train_model(pretrained=True, train_dir=dir, val_dir=validation_dir) 
    new_resnet50_pretrained.append(val_acc_pretrained)

    print(f"Training baseline model with {size} samples...")
    val_acc_baseline = resnet_train_model(pretrained=False, train_dir=dir, val_dir=validation_dir)
    new_resnet50_baseline.append(val_acc_baseline)
    
    print("\npretrained array: ", new_resnet50_pretrained, "\nbaseline array: ", new_resnet50_baseline)
```

```python
from keras.applications import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout
from keras import optimizers
import numpy as np

# Function to create and train a model
def train_vgg16_model(pretrained=True, train_dir = './setA/train_dir', val_dir='./validation_dir', training_size = 20):
    if pretrained:
        conv_base = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
        conv_base.trainable = False
    else:
        conv_base = VGG16(weights=None, include_top=False, input_shape=(224, 224, 3))
        conv_base.trainable = True  # Train the whole model if no pretrained weights are used

    # Preprocessing the data - only rescaling the images.
    datagen = ImageDataGenerator(rescale=1./255)
    
    # Data generators
    train_generator = datagen.flow_from_directory(
        train_dir,
        target_size=(224, 224),
        batch_size=20,
        class_mode='binary')

    validation_generator = datagen.flow_from_directory(
        val_dir,
        target_size=(224, 224),
        batch_size=20,
        class_mode='binary')

    # Create the model and add new dense layers for binary classification.
    model = Sequential()
    model.add(conv_base)
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(optimizer=optimizers.RMSprop(learning_rate=2e-5),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=20,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size)
    
    # Return the best validation accuracy
    return max(history.history['val_accuracy'])

# Training sizes and directories setup
new_training_sizes = [20, 50, 90, 120, 180, 250]
new_training_dirs = ['./setA/train_dir', './setB/train_dir', './setC/train_dir', './setD/train_dir', './setE/train_dir', './setF/train_dir']
validation_dir = './validation_dir'  # Fixed validation directory
new_vgg16_pretrained = []
new_vgg16_baseline = []

# Train models for each size
for size, dir in zip(new_training_sizes, new_training_dirs):
    print(f"Training pretrained VGG16 model with {size} samples...")
    val_acc_pretrained = train_vgg16_model(pretrained=True, train_dir=dir, val_dir=validation_dir, training_size = size)
    new_vgg16_pretrained.append(val_acc_pretrained)

    print(f"Training baseline VGG16 model with {size} samples...")
    val_acc_baseline = train_vgg16_model(pretrained=False, train_dir=dir, val_dir=validation_dir, training_size = size)
    new_vgg16_baseline.append(val_acc_baseline)

    print("\npretrained array: ", new_vgg16_pretrained, "\nbaseline array: ", new_vgg16_baseline)
```
